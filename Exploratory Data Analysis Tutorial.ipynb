# Step 2: Data Preperation

- Dropping irrelevant columns and rows
- Identifying duplicated columns
- Renaming columns
- Feature creation

```python
# Dropping columns
df = df[['coaster_name', 
     # 'Length', 'Speed', 
     'Location', 'Status', 
     # 'Opening date', 'Type', 
     'Manufacturer', 
     # 'Height restriction', 'Model', 'Height', 'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
     # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle', 'Drop', 'Soft opening date', 
     # 'Fast Lane available', 'Replaced', 'Track layout', 'Fastrack available', 'Soft opening date.1', 'Closing date', 'Opened', 
     # 'Replaced by', 'Website', 'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
     # 'Single rider line available', 'Restraint Style', 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
     'year_introduced', 
     'latitude', 'longitude', 
     'Type_Main',
     'opening_date_clean', 
     # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
     'speed_mph', 
     # 'height_value', 'height_unit', 
     'height_ft',
     'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.shape
```

```python
df.dtypes 
```

```python
# Changed date type from object to datetime 
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean']) 
```

```python
# Rename our columns
df = df.rename(columns={'coaster_name':'Coaster_Name',
                  'year_introduced':'Year_Introduced',
                  'opening_date_clean':'Opening_Date',
                  'speed_mph':'Speed_mph',
                  'height_ft':'Height_ft',
                  'Inversions_clean':'Inversions',
                  'Gforce_clean':'Gforce',
                  'latitude':'Latitude',
                  'longitude':'Longitude'}).copy()
```

```python
# Identify missing and null values
df.isna().sum()
```

```python
df.loc[df.duplicated(subset=['Coaster_Name'])]
```

# Checking an example duplicate
df.query('Coaster_Name == "Crystal Beach Cyclone"')

```python
# Seeing total number of duplicates
df.duplicated(subset=['Coaster_Name','Location','Opening_Date']).sum()
```

```python
# Removing duplicates
df = df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])] \
    .reset_index(drop=True).copy()
```

# Step 4: Feature Relationships

- Scatterplot
- Heatmap correlation
- Pairplot
- Groupby comparisons

```python
# Compare 2 features
df.plot(kind='scatter', 
        x='Speed_mph', 
        y='Height_ft',
        title='Coaster Speed vs. Height')
plt.show()
```

```python
# Compare 2 features with hue and data key
sns.scatterplot(x='Speed_mph', 
                y='Height_ft',
                hue='Year_Introduced',
                data=df)
plt.show()
```

```python
# Compare more than 2 features, with hue, and data key
import warnings
sns.pairplot(df, vars=['Year_Introduced','Speed_mph','Height_ft','Inversions','Gforce'],
            hue='Type_Main')
warnings.simplefilter(action='ignore')
plt.show()
```

```python
# Correlation between different features
df_corr = df[['Year_Introduced','Speed_mph','Height_ft','Inversions','Gforce']].dropna().corr()
df_corr
```

```python
# Heatmap of correlations
sns.heatmap(df_corr, annot=True)
plt.show()

# Might be seaborn version issue causing this first row problem, do 'pip install seaborn --upgrade'
```

# Step 5: Ask a Question About the Data

Try to answer a question you have about the data using a plot or statistic:
- What are the locations with the fastest roller coasters (minimum of 10)?

```python
df['Location'].value_counts()
```

```python
# agg means aggregation
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coast Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
